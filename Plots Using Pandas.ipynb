# Line Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data
data = {'year': [2010, 2011, 2012, 2013, 2014],
        'sales': [5, 7, 3, 8, 10]}
df = pd.DataFrame(data)

plt.figure(figsize=(4, 3))
df.plot(x='year', y='sales', kind='line')
plt.title('Sales Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

# Area Plot

```python
# Example data
data = {'A': [1, 3, 2, 5],
        'B': [4, 2, 3, 6]}
df = pd.DataFrame(data)

# Plotting
df.plot(kind='area')
plt.title('Area Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

# Histogram

```python
# Example data
data = {'age': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]}
df = pd.DataFrame(data)

# Plotting
df['age'].plot(kind='hist', bins=5)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# Bar Chart

```python
# Example data
data = {'category': ['A', 'B', 'C'],
        'values': [10, 20, 15]}
df = pd.DataFrame(data)

# Plotting
df.plot(x='category', y='values', kind='bar')
plt.title('Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

# Pie Chart

```python
# Example data
data = {'percentage': [30, 20, 50]}
df = pd.DataFrame(data, index=['A', 'B', 'C'])

# Plotting
df.plot(y='percentage', kind='pie', autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.ylabel('')
plt.show()
```

# Box Plot

```python
# Example data
data = {'values': [5, 10, 15, 20, 25]}
df = pd.DataFrame(data)

# Plotting
df.plot(kind='box')
plt.title('Box Plot Example')
plt.ylabel('Values')
plt.show()
```

# Scatter Plot

```python
# Example data
data = {'height': [150, 160, 170, 180, 190],
        'weight': [50, 55, 65, 70, 80]}
df = pd.DataFrame(data)

# Plotting
df.plot(x='height', y='weight', kind='scatter')
plt.title('Scatter Plot Example')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```
